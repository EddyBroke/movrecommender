```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Keeping only theese columns 
#id
#keywords
#overview
#genre
#title
#cast
#crew
movies = movies[['movie_id','keywords','overview','genres','title','cast','crew']]
```

```python
movies.head(30)
```

```python
#check missing data values in each columns
movies.isnull().sum()
```

```python
#Dropping the null values...
movies.dropna(inplace = True)
```

```python
movies.iloc[1].genres
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# getting 3 major cast of the movie
def convert3(obj):
    L = []
    counter = 0;
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter = counter +1 ;
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#getting the directors of the movies

def convert4(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert4)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) # converting each element to seperate element
```

```python
movies.head()
```

```python
#removing spaces
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ', '')for i in x ])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ', '')for i in x ])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ', '')for i in x ])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ', '')for i in x ])
```

```python
movies.head()
```

```python
#combining the columns
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#creating new dataframe
new_df = movies[['movie_id' , 'title', 'tags']]
```

```python
new_df.head()
```

```python
#converting list to string...
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
# converting everything to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
# now we will be converting tags column to vector
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
# using cv object to fit transform and converting it to numpy array as count vectorizer gives scipy parse matrix wit 0's
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split(): # converting list element to string
        y.append(ps.stem(i))
    return " ".join(y) # converting back string to list element
```

```python
ps.stem('loving')
```

```python
# we are using nltk to make sure the words which are similar to each other like loving love loved may get a imilar use case word such as love only....
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# in higher dimension eucladein distance measurement is not an accurate manner to measure because it givs points point distance not wiht respect to the cosine angle ...
```

```python
from sklearn.metrics.pairwise import cosine_similarity # we took similarity becasue 1 means more similarity 0 means lesss similarity
```

```python
similarity = cosine_similarity(vectors) # passing all the vectors to calculate the distance betweent them ....
```

```python
similarity[0]# shows distance of first movie(vector form) with every other movie (vector form)
```

```python
new_df[new_df['title'] == ''] # find movie here
```

```python
# finding the similar movie now
# we will get the index of the movie
# then go into the similarity matrix's that index where the actual movie provided by user resides
# after going there we would get its distance from every other movie and then
# we will sort it in ascending order (also keep its index in hand so that new order may not hamper with 0th movie distance with 1st movie distance) and display top 5 or any number of movies 
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6] # selecting top 7 movies similar to the user's choice and also applying ascending order to it while keeping in mind the index.
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title) # will give the title of the movie 
    
    
```

```python
recommend('Pixels')
```
